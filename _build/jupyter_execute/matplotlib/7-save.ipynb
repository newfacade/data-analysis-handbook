# 保存和加载图片

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
# matplotlib加入中文支持
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']
```

## 保存图片

```python
plt.plot(np.random.rand(10), np.random.rand(10))
"""
保存图片
dpi指示图片分辨率
可保存为png、svg(矢量)、pdf(矢量)等格式
"""
plt.savefig("save_sample.svg", dpi=300, format="svg")
```

## 加载图片

```python
img = plt.imread("./line_styles_reference.png") # 不支持加载svg格式
plt.imshow(img)
plt.show()
```
