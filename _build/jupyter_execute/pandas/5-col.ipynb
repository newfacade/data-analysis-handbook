# 变更行和列

```python
import pandas as pd
import numpy as np
```

## 修改行和列的名称

```python
df = pd.read_csv('sample.csv', index_col=0)
df
```

```python
df = df[df['up'] != 75]
df
```

```python
df.columns
```

```python
# 可直接修改列名，行也是一样
df.columns = np.arange(6)
df
```

## 修改行、列的顺序

```python
# 使用reindex修改行的顺序
df.reindex(['c', 'd', 'e', 'f', 'g', 'h', 'xx', 'yy'])
```

```python
# 直接修改列的顺序
df[[5, 4, 3, 2, 1, 0]]
```

## 改变索引

```python
# 使用reset_index重置索引
df.reset_index(drop=False)
```

```python
# 使用set_index用现有的列做索引
df.set_index(keys=[0, 1])
```
