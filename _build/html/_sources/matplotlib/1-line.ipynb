# 线图

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']  # matplotlib加入中文支持
```

## 基本线图

```python
x = np.arange(1, 6)
```

color: 颜色

linestyle: 线型

linewidth: 线宽

marker: 标记类型

markersize: 标记大小

label: 名字

```python
kwargs = dict(color='green', linestyle='--', linewidth=2, marker='o', markersize=12)
```

```python
plt.plot(x, np.pi * x ** 2, **kwargs, label='fast')
"""
让图完整起来
"""
plt.xlim(1, 5)
plt.ylim(0, 80)
plt.xticks(np.arange(1.0, 5.0, 0.5))
plt.yticks(np.arange(0, 80, 5))
plt.xlabel("半径")
plt.ylabel("面积")
plt.title("圆的半径和面积")
plt.legend(loc="upper left")
plt.grid()
```

## 多子图

```python
fig, ax = plt.subplots(2)
t = np.linspace(0, 10, 100)
ax[0].plot(t, np.sin(t))
ax[1].plot(t, np.cos(t))
```
